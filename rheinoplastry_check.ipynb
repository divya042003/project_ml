```python
import os
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import cv2
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from google.colab import files
import shutil
```

```python
before_path = "/content/before"
after_path = "/content/after"

os.makedirs(before_path, exist_ok=True)
os.makedirs(after_path, exist_ok=True)

print("Folders created! Now upload images.")
```

```python
IMG_SIZE = (128, 128)
BATCH_SIZE = 16

datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

train_data = datagen.flow_from_directory(
    "/content/",
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary',
    subset="training"
)

val_data = datagen.flow_from_directory(
    "/content/",
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary',
    subset="validation"
)
```

```python
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(128, 128, 3)),
    MaxPooling2D(2,2),

    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),

    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.3),
    Dense(1, activation='sigmoid')  # Binary classification
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.summary()
```

```python
def predict_image(image_path):
    img = cv2.imread(image_path)
    img = cv2.resize(img, IMG_SIZE) / 255.0
    img = np.expand_dims(img, axis=0)

    prediction = model.predict(img)
    return "After Surgery" if prediction > 0.5 else "Before Surgery"

# Upload and predict
uploaded_file = files.upload()
for filename in uploaded_file.keys():
    result = predict_image(filename)
    print(f"Prediction for {filename}: {result}")
```
